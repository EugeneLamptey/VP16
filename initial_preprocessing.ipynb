```python
#Importing the necessry libraries
import pandas as pd
```

```python
path_all_data = "Datasets\AID_651615_datatable_all.csv"
data = pd.read_csv(path_all_data)
data.head(7)
```

```python
#removing irrelevant rows
data = data.iloc[4:, :]
data.info()
```

Below is a preprocessor to filter out missing entries, duplicated entries and irrelevant columns

```python
print('the number of duplicated compounds: ', data['PUBCHEM_CID'].duplicated().sum())
```

```python
relevant_cols = {'PUBCHEM_CID': 'Compound_Id',
                'PUBCHEM_EXT_DATASOURCE_SMILES': 'Canonical_smiles',
                  'PUBCHEM_ACTIVITY_OUTCOME': 'Activity'}

#Relevant cols is of type dict
#preprocess removes duplicates and null values
def preprocess(data, relevant_cols):
    df = data[list(relevant_cols.keys())].copy()
    df.columns = df.columns.map(relevant_cols)
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True, ignore_index=True)
    return df
```

```python
preprocessed_result = preprocess(data, relevant_cols)
preprocessed_result['Activity'] = preprocessed_result['Activity'].map(lambda x: 1 if x=='Active' else 0)

preprocessed_result
```
